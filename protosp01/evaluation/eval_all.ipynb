```python
import pandas as pd
from evaluate_src import *
import os
```

```python
res_df = pd.DataFrame(columns=['path','seqeval_precision', 'seqeval_recall', 'seqeval_f1', 'skillevel_precision', 'skillevel_recall', 'skillevel_f1', 'filtered_skillevel_precision', 'filtered_skillevel_recall', 'filtered_skillevel_f1', 'filtered_seqeval_precision', 'filtered_seqeval_recall', 'filtered_seqeval_f1'])
for shots in [5]:
    for model in ['gpt-3.5-turbo']:
        for prompt_type in ['ner', 'extract']:
            for dataset_name in ['green', 'skillspan', 'fijo', 'sayfullina', 'kompetencer', 'gnehm']:
                save_path = 'output/' + model + '/' + dataset_name + '_' + prompt_type + '_' + str(shots) + '-shots_knn.json'
                if not os.path.exists(save_path):
                    continue
                all_metrics = eval(save_path)
                res_df = pd.concat([res_df, pd.DataFrame([{'path': save_path, **all_metrics}])], ignore_index=True)
```
