```python
import pandas as pd
from evaluate_src import *
import os
```

```python
res_df = pd.DataFrame(columns=['path', 'precision', 'recall', 'f1', 'skill_level_accuracy', 'filtered_precision', 'filtered_recall', 'filtered_f1', 'filtered_skill_level_accuracy'])
for shots in [4]:
    for model in ['gpt-3.5-turbo']:
        for prompt_type in ['ner', 'extract']:
            for dataset_name in ['gnehm', 'green', 'skillspan', 'fijo', 'sayfullina', 'kompetencer']:
                save_path = 'output/' + dataset_name + '_' + model + '_' + prompt_type + '_' + str(shots) + '-shots.json'
                if not os.path.exists(save_path):
                    continue
                res_dict, res_dict_filtered = eval(save_path)
                res_dict_filtered = {'filtered_' + k : v for k, v in res_dict_filtered.items()}
                res_df = res_df.append({'path': save_path, **res_dict, **res_dict_filtered}, ignore_index=True)

```
